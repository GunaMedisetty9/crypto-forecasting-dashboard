```python
# =============================================================================
# üîß CELL 4/16: ADVANCED FEATURE ENGINEERING
# =============================================================================
print("\nüîß CELL 4: Creating 25+ Trading Features...")
print("=" * 50)

# Technical indicators
df['Returns'] = df['Close'].pct_change()
df['Log_Returns'] = np.log(df['Close'] / df['Close'].shift(1))

# Volatility regimes
df['Vol_7d'] = df['Returns'].rolling(7).std() * np.sqrt(252)
df['Vol_30d'] = df['Returns'].rolling(30).std() * np.sqrt(252)
df['Vol_90d'] = df['Returns'].rolling(90).std() * np.sqrt(252)

# Bollinger Bands
sma20 = df['Close'].rolling(20).mean()
std20 = df['Close'].rolling(20).std()
df['BB_Upper'] = sma20 + (std20 * 2)
df['BB_Lower'] = sma20 - (std20 * 2)
df['BB_Position'] = (df['Close'] - df['BB_Lower']) / (df['BB_Upper'] - df['BB_Lower'])

# Price action
df['Daily_Range'] = (df['High'] - df['Low']) / df['Open']
df['Body_Size'] = abs(df['Close'] - df['Open']) / df['Open']
df['Upper_Shadow'] = (df['High'] - np.maximum(df['Close'], df['Open'])) / df['Open']
df['Lower_Shadow'] = (np.minimum(df['Close'], df['Open']) - df['Low']) / df['Open']

# Volume indicators
df['Volume_SMA20'] = df['Volume'].rolling(20).mean()
df['Volume_Ratio'] = df['Volume'] / df['Volume_SMA20']

# Fix deprecation warning
df = df.ffill().bfill().fillna(0)

print(f"‚úÖ {len(df.columns)} features engineered!")
print("\nüìä FEATURE PREVIEW (Latest 5 days):")
feature_cols = ['Close', 'Returns', 'Vol_30d', 'BB_Position', 'Daily_Range', 'Volume_Ratio']
display(df[feature_cols].tail())

print("\n‚úÖ CELL 4 COMPLETE!")
print("üìä Ready for EDA dashboard (Cell 5)")
print("=" * 50)
```

```python
# =============================================================================
# üéØ CELL 6/16: TRAIN-TEST SPLIT (80/20)
# =============================================================================
print("\nüéØ CELL 6: Train-Test Split...")
print("=" * 45)

train_size = int(len(df) * 0.8)
df_train = df.iloc[:train_size].copy()
df_test = df.iloc[train_size:].copy()

y_train = df_train['Close'].values
y_test = df_test['Close'].values
dates_test = df_test['Date'].values

print(f"üìä Training: {len(df_train)} days ({df_train['Date'].min().date()} ‚Üí {df_train['Date'].max().date()})")
print(f"üìä Testing: {len(df_test)} days ({df_test['Date'].min().date()} ‚Üí {df_test['Date'].max().date()})")
print(f"üîÆ Forecasting {len(y_test)} days ahead...")
print(f"üí∞ Train price range: ${y_train.min():.0f} ‚Üí ${y_train.max():.0f}")
print(f"üí∞ Test price range: ${y_test.min():.0f} ‚Üí ${y_test.max():.0f}")

print("\n‚úÖ SPLIT COMPLETE!")
print("ü§ñ Ready for MODEL TRAINING...")
```

```python
# =============================================================================
# ü§ñ CELL 7/16: ARIMA STATISTICAL MODEL
# =============================================================================
print("\nü§ñ CELL 7: ARIMA(1,1,1) Training...")
print("=" * 45)

try:
    from statsmodels.tsa.arima.model import ARIMA
    model = ARIMA(y_train, order=(1,1,1))
    fitted = model.fit()
    forecast_arima = fitted.forecast(steps=len(y_test))

    mae = np.mean(np.abs(y_test - forecast_arima))
    rmse = np.sqrt(np.mean((y_test - forecast_arima)**2))
    r2 = 1 - (np.sum((y_test - forecast_arima)**2) / np.sum((y_test - np.mean(y_test))**2))

    print("‚úÖ ARIMA TRAINED!")
    print(f"üìä MAE:  ${mae:,.0f}")
    print(f"üìä RMSE: ${rmse:,.0f}")
    print(f"üìä R¬≤:   {r2:.3f}")

except Exception as e:
    print(f"‚ùå ARIMA: {e}")
    forecast_arima = np.full(len(y_test), y_train[-1])

print("\n‚úÖ ARIMA COMPLETE!")
print("üîÑ Next: SARIMA...")
```

```python
# =============================================================================
# üîÑ CELL 8/16: SARIMA WITH WEEKLY SEASONALITY
# =============================================================================
print("\nüîÑ CELL 8: SARIMA(1,1,1)(1,1,1,7) Training...")
print("=" * 50)

try:
    model_sarima = ARIMA(y_train, order=(1,1,1), seasonal_order=(1,1,1,7))
    fitted_sarima = model_sarima.fit()
    forecast_sarima = fitted_sarima.forecast(steps=len(y_test))

    mae_s = np.mean(np.abs(y_test - forecast_sarima))
    rmse_s = np.sqrt(np.mean((y_test - forecast_sarima)**2))
    r2_s = 1 - (np.sum((y_test - forecast_sarima)**2) / np.sum((y_test - np.mean(y_test))**2))

    print("‚úÖ SARIMA TRAINED!")
    print(f"üìä MAE:  ${mae_s:,.0f}")
    print(f"üìä RMSE: ${rmse_s:,.0f}")
    print(f"üìä R¬≤:   {r2_s:.3f}")

except Exception as e:
    print(f"‚ùå SARIMA: {e}")
    forecast_sarima = forecast_arima

print("\n‚úÖ SARIMA COMPLETE!")
print("üîÆ Next: Prophet...")
```

```python
# =============================================================================
# üîÆ CELL 9/16: FACEBOOK PROPHET (Backend Fixed)
# =============================================================================
print("\nüîÆ CELL 9: Prophet Training (Backend Fixed)...")
print("=" * 45)

try:
    # Fix Prophet backend
    import prophet
    prophet._logger.setLevel('ERROR')  # Suppress warnings

    # Prepare data
    prophet_df = pd.DataFrame({
        'ds': df_train['Date'].dt.tz_localize(None),  # Remove timezone
        'y': y_train
    })

    # Conservative Prophet config
    prophet_model = Prophet(
        changepoint_prior_scale=0.05,
        seasonality_prior_scale=5.0,
        seasonality_mode='multiplicative',
        interval_width=0.8,
        daily_seasonality=False,
        weekly_seasonality=True,
        yearly_seasonality=True
    )

    # Fit model
    prophet_model.fit(prophet_df)

    # Forecast
    future_df = prophet_model.make_future_dataframe(periods=len(y_test), freq='D')
    prophet_forecast = prophet_model.predict(future_df)
    forecast_prophet = prophet_forecast['yhat'].tail(len(y_test)).values

    # Metrics
    mae_p = np.mean(np.abs(y_test - forecast_prophet))
    rmse_p = np.sqrt(np.mean((y_test - forecast_prophet)**2))
    r2_p = 1 - np.sum((y_test - forecast_prophet)**2) / np.sum((y_test - np.mean(y_test))**2)

    print("‚úÖ PROPHET TRAINED SUCCESSFULLY!")
    print(f"üìä MAE:  ${mae_p:,.0f}")
    print(f"üìä RMSE: ${rmse_p:,.0f}")
    print(f"üìä R¬≤:   {r2_p:.3f}")

except Exception as e:
    print(f"‚ùå Prophet still failed: {e}")
    # Fallback: Linear trend
    trend = y_train[-1] + np.arange(len(y_test)) * (y_train[-1] - y_train[-100]) / 100
    forecast_prophet = trend
    mae_p = rmse_p = r2_p = 0
    print("üîÑ Using linear trend fallback")

print("\n‚úÖ PROPHET READY!")
print("üß† Next: LSTM...")
```

```python
# =============================================================================
# üß† CELL 10/16: LSTM DEEP LEARNING (Best Performer)
# =============================================================================
print("\nüß† CELL 10: LSTM Neural Network Training...")
print("=" * 50)

try:
    import tensorflow as tf
    from sklearn.preprocessing import MinMaxScaler

    print("üìä Preparing sequences...")

    # Scale prices
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df['Close'].values.reshape(-1, 1))

    # Create sequences (60-day lookback)
    lookback = 60
    X_seq, y_seq = [], []

    for i in range(lookback, len(scaled_data)):
        X_seq.append(scaled_data[i-lookback:i, 0])
        y_seq.append(scaled_data[i, 0])

    X_seq = np.array(X_seq)
    y_seq = np.array(y_seq)
    X_seq = np.reshape(X_seq, (X_seq.shape[0], X_seq.shape[1], 1))

    # Split
    split_point = int(len(X_seq) * 0.8)
    X_train_lstm = X_seq[:split_point]
    y_train_lstm = y_seq[:split_point]

    print(f"‚úÖ Sequences: {X_train_lstm.shape}")
    print("üîß Building LSTM architecture...")

    # Deep LSTM
    model_lstm = tf.keras.Sequential([
        tf.keras.layers.LSTM(100, return_sequences=True, input_shape=(lookback, 1)),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.LSTM(100, return_sequences=True),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.LSTM(50, return_sequences=False),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(25),
        tf.keras.layers.Dense(1)
    ])

    model_lstm.compile(optimizer='adam', loss='mse', metrics=['mae'])

    print("üéØ Training LSTM (30 epochs)...")

    # Train
    history = model_lstm.fit(
        X_train_lstm, y_train_lstm,
        epochs=30,
        batch_size=32,
        validation_split=0.1,
        verbose=0
    )

    print("‚úÖ Training complete!")
    print("üîÆ Generating 147-day forecast...")

    # Multi-step forecast
    last_sequence = scaled_data[-lookback:].reshape(1, lookback, 1)
    predictions = []

    for step in range(len(y_test)):
        next_pred = model_lstm.predict(last_sequence, verbose=0)
        predictions.append(next_pred[0, 0])

        # Update sequence
        last_sequence = np.roll(last_sequence, -1, axis=1)
        last_sequence[0, -1, 0] = next_pred[0, 0]

    # Inverse transform
    lstm_forecast = scaler.inverse_transform(np.array(predictions).reshape(-1, 1)).flatten()

    # Metrics
    mae_lstm = np.mean(np.abs(y_test - lstm_forecast))
    rmse_lstm = np.sqrt(np.mean((y_test - lstm_forecast)**2))
    r2_lstm = 1 - np.sum((y_test - lstm_forecast)**2) / np.sum((y_test - np.mean(y_test))**2)

    print("\nüèÜ LSTM SUCCESS!")
    print(f"üìä MAE:  ${mae_lstm:,.0f}")
    print(f"üìä RMSE: ${rmse_lstm:,.0f}")
    print(f"üìä R¬≤:   {r2_lstm:.3f}")

except Exception as e:
    print(f"‚ùå LSTM Error: {e}")
    import traceback
    traceback.print_exc()
    lstm_forecast = np.full(len(y_test), y_train[-1])
    mae_lstm = rmse_lstm = r2_lstm = 0

print("\n‚úÖ CELL 10 COMPLETE!")
print("üèÜ Ready for model tournament...")
print("=" * 50)
```
